```python
columns = ['CO2_ppm', 'Humidity_%','Pressure_hPa',
       'WindSpeed_km_h', 'PollutionIndex', 'Temperature_C', 'HeatIndex',
       'Rainfall_mm']
df[columns] = df[columns].astype('int64')
```

```python
df.dtypes
```

Managing **Object**  Data type

```python
def convert_obj_into_int(columns):
    for i in columns:
        data = pd.DataFrame({i: df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        df[i] = data_LE.transform(df[i])

columns = ['Region', 'Season']
convert_obj_into_int(columns)
```
