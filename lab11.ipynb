```python
columns = ['CO2_ppm', 'Humidity_%','Pressure_hPa',
       'WindSpeed_km_h', 'PollutionIndex', 'Temperature_C', 'HeatIndex',
       'Rainfall_mm']
df[columns] = df[columns].astype('int64')
```

```python
df.dtypes
```

Managing **Object**  Data type

```python
def convert_obj_into_int(columns):
    for i in columns:
        data = pd.DataFrame({i: df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(data))
        df[i] = data_LE.transform(df[i])

columns = ['Region', 'Season']
convert_obj_into_int(columns)
```

**Train** - **Test Splitting**

```python
df.columns
```

```python
x = df.drop('Temperature_C',axis=1)
y = df['Temperature_C']
```

```python
train_X, test_X, train_y, test_y = train_test_split(x, y, test_size=0.2, shuffle=False)
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

**Model Training**

```python
model_svr = SVR()
model_svr.fit(train_X, train_y)

print(model_svr)
```

```python
pred_svr = model_svr.predict(test_X)
print(pred_svr)
```

```python
# Predictions
pred_svr = model_svr.predict(test_X)

# R² Score as percentage
r2_percentage = round(r2_score(test_y, pred_svr) * 100, 2)
print(f"Model R²: {r2_percentage} %")
```

**Model Saving** and **Model Loading**

```python
pickle.dump(model_svr, open('model_svr.pkl', 'wb'))   # Model Saving 
```
